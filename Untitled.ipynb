```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

plesnipari = pd.read_csv('plesnipari.csv', index_col=['idplesalca', 'drzava', 'plesalec', 'plesalka', 'kategorija'])
pd.options.display.max_rows = 20
plesnipari.head(10)

plesnipari=pd.read_csv("plesnipari.csv",parse_dates=['drzava'])
plesnipari.head()
plt.rcParams['figure.figsize'] = 16, 12
```

```python
plesnipari['drzava'].describe()  
## koliko je držav in iz katere jih je največ
```

```python
plesnipari['kategorija'].describe()   
## koliko kategorij in katera je najpogostejša
```

```python
kategorija = plesnipari[['idplesalca', 'kategorija', 'plesalec']]
kategorija.head()
kategorija= kategorija.groupby('kategorija')
kategorija.size().plot(kind='bar')                               
##  graf prikazuje koliko je tekmovalcev v kateri kategoriji
```

```python
drzave = plesnipari[['idplesalca', 'drzava', 'plesalec']]
drzave.head()
drzave= drzave.groupby('drzava')
drzave.size().plot(kind='bar')           
## graf prikazuje koliko tekmovalcev je iz katere države
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

tekmovanja = pd.read_csv('tekmovanja.csv')
pd.options.display.max_rows = 20
tekmovanja.head(10)                      
##  prikaže tabelo 'tekmovanja' in sicer prvih 10 vrst
```

```python
tekmovanja = pd.read_csv('tekmovanja.csv')
tekmovalci = tekmovanja.groupby('ime')
tekmovalci.mean()     
## izračuna povprečje uvrstitve tekmovalcev
```

```python
tekmovanja.sort_values(['ime', 'uvrstitev'], ascending=[True, True])
##  izpiše tebelo 'tekmovanja' in jo uredi po abecednem redu imena, če je ime enako uredi po uvrstitvi
```

```python
plt.rcParams['figure.figsize'] = 16, 12
dogodek = tekmovanja[['ime', 'dogodek', 'datum']]
dogodek.head()
dogodek= dogodek.groupby('dogodek')
dogodek.size().plot(kind='bar')
##  graf prikazuje katere vrste je bilo tekmovanje
```

```python
drzava2 = tekmovanja[['drzava2', 'dogodek', 'datum']]
drzava2.head()
drzava2= drzava2.groupby('drzava2')
drzava2.size().plot(kind='bar')
##prikazuje iz koliko je bilo tekmovalcev v posamezni državi
```

```python
disciplina = tekmovanja[['disciplina', 'dogodek', 'datum']]
disciplina.head()
disciplina= disciplina.groupby('disciplina')
disciplina.size().plot(kind='bar')
##  prikazuje koliko tekmovanj je bilo prirejenih glede na disciplino
```
